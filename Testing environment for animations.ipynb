Start by importing all the neccessary Python packages and setting some base parameters

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
rc('animation', html='jshtml')
```

Initialize the figure and helped function for future plotting

```python
fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
def init():
    line.set_data([], [])
    return (line,)
```

Testing function for animations

```python
def animate_test1(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)
```

Set the plot generation function

```python
ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 3))
anim = animation.FuncAnimation(fig, animate_test1, init_func=init,
                               frames=100, interval=20, 
                               blit=True)
```
